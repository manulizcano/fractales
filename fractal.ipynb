```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
from random import uniform
from mpl_toolkits.mplot3d import axes3d
import plotly.express as px
import pandas as pd
```

```python
def fractal(n_iteraciones):
  a = np.array([0,0])
  b = np.array([1,0])
  c = np.array([1/2, (3**(1/2))/2])
  posibilidades = {0:a, 1:b, 2:c}

  al = int(uniform(0,3))
  inicio = posibilidades[al]
  x=[0,1,1/2]
  y=[0,0,(3**(1/2))/2]

  for i in range(n_iteraciones):
    rand = int(uniform(0,3))
    cambio = posibilidades[rand]
    inicio =  inicio + cambio
    inicio = inicio/2
    x.append(inicio[0])
    y.append(inicio[1])
  fig = plt.figure(figsize=(9, 9), dpi=80)
  ax =plt.subplot()
  ax.scatter(x,y)
  plt.show()
```

```python
fractal(100000)
```

```python
def fractal_3d(n_iteraciones):
  a = np.array([0,0,0])
  b = np.array([1,0,0])
  c = np.array([1/2, (3**(1/2))/2,0])
  d = np.array([1/2,(3**(1/2))/6,(2/3)**(1/2)])
  posibilidades = {0:a, 1:b, 2:c,3:d}

  al = int(uniform(0,4))
  inicio = posibilidades[al]
  x=[a[0],b[0],c[0],d[0]]
  y=[a[1],b[1],c[1],d[1]]
  z=[a[2],b[2],c[2],d[2]]
  for i in range(n_iteraciones):
    rand = int(uniform(0,4))
    cambio = posibilidades[rand]
    inicio =  inicio + cambio
    inicio = inicio/2
    x.append(inicio[0])
    y.append(inicio[1])
    z.append(inicio[2])

  datos = {'x': x, 'y':y, 'z':z}
  df = pd.DataFrame(data = datos)
  df['tam'] = np.zeros(len(df)) + 0.000001
  fig = px.scatter_3d(df, x='x', y='y', z='z',size='tam', opacity=1)
  fig.show()
```

```python
fractal_3d(50000)
```
